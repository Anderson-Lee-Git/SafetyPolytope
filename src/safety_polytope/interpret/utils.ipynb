```python
def find_weights_paths(base_dir):
    run_summaries = find_run_summaries(base_dir)
    weights_paths = []
    for summary in run_summaries:
        with open(summary, "r") as f:
            data = json.load(f)
        weights_path = data.get("weights_save_path", None)
        if weights_path:
            weights_paths.append(weights_path)
    return weights_paths


base_dir = "/scratch/gpfs/KOROLOVA/cl6486/SafetyPolytope/multirun/beaver_tails/2025-10-17/polytope_hb-22-54-56"

for p in find_run_summaries(base_dir):
    print(p, end=",")
```
